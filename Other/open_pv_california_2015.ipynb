```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
pv_data = "open_pv_california_2015.csv"

# Read the City and Ride Data
pv_data_df = pd.read_csv(pv_data)

df = pv_data_df.rename(columns = {"size_kw":"system_size_kw","annual_PV_prod":"annual_PV_prod_kwh"})
# change size_kw to system_size_kw, annual_PV_prod_kwh
df.head()
```

```python
df2 = df[["state","date_installed","incentive_prog_names","system_size_kw","zipcode","install_type",
                                                "installer","cost_per_watt","cost","utility_clean","county","annual_PV_prod_kwh"]]

df3 = df2.dropna()
df3

```

```python
# File to Load (Remember to change these)
climate_zip_code = "BuildingClimateZonesByZIPCode.csv"
population = "population.csv"

# Read the data
climate_zip_code_df = pd.read_csv(climate_zip_code)
population_df = pd.read_csv(population)

# Combine the data into a single dataset
combined_data_df_2 = pd.merge(df3, climate_zip_code_df, on="zipcode")

combined_data_df_3 = pd.merge(combined_data_df_2, population_df, on="county")

# # Display the data table for preview
combined_data_df_3

```

```python
install_type = combined_data_df_3.set_index("install_type")

# Create DF for residential
residential_df = install_type.loc[["residential"],["state","date_installed","incentive_prog_names","system_size_kw","zipcode","installer","cost_per_watt","cost","utility_clean","county","annual_PV_prod_kwh","Building CZ","population","housing_units","land_area_sq_mi"]]

# Save as a csv
residential_df.to_csv("residential_data.csv", encoding="utf-8", index=False)
residential_df
```
